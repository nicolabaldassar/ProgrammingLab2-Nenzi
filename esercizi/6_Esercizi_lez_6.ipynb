### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")

###

# Funzione coseno stagionale
def seasonal_cosine(x, A, B, C, D):
    return A * np.cos(B * x + C) + D

# Fit per temp_max
popt_max, _ = sp.optimize.curve_fit(seasonal_cosine, months, temp_max, p0=[10, 2*np.pi/12, 0, 25])
temp_max_fit = seasonal_cosine(months, *popt_max)
mae_max = np.mean(np.abs(temp_max - temp_max_fit))
rmse_max = np.sqrt(np.mean((temp_max - temp_max_fit) ** 2))

# Fit per temp_min
popt_min, _ = sp.optimize.curve_fit(seasonal_cosine, months, temp_min, p0=[30, 2*np.pi/12, 0, -30])
temp_min_fit = seasonal_cosine(months, *popt_min)
mae_min = np.mean(np.abs(temp_min - temp_min_fit))
rmse_min = np.sqrt(np.mean((temp_min - temp_min_fit) ** 2))

# Plot
plt.plot(months, temp_max, "ro", label="Temp max")
plt.plot(months, temp_max_fit, "r--", label=f"Fit max (MAE={mae_max:.1f}, RMSE={rmse_max:.1f})")
plt.plot(months, temp_min, "bo", label="Temp min")
plt.plot(months, temp_min_fit, "b--", label=f"Fit min (MAE={mae_min:.1f}, RMSE={rmse_min:.1f})")
plt.xlabel("Mese")
plt.ylabel("Temperatura (°C)")
plt.title("Fit stagionale su temperature massime e minime")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
# Seleziona variabili
X = df['disp'].values
Y = df['mpg'].values

# Calcola regressione lineare
slope, intercept, r_value, p_value, std_err = sp.stats.linregress(X, Y)

print(f"Equazione retta: mpg = {slope:.4f} * disp + {intercept:.4f}")
print(f"R-quadrato: {r_value**2:.4f}")

# Predizioni
Y_pred = slope * X + intercept

# Plot dati e retta
plt.scatter(X, Y, label='Dati originali')
plt.plot(X, Y_pred, color='red', label='Fit lineare')
plt.xlabel('Displacement (disp)')
plt.ylabel('Miles per Gallon (mpg)')
plt.title('Regressione lineare tra disp e mpg')
plt.legend()
plt.grid(True)
plt.show()
```
